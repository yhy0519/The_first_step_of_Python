코딩예제3. 위에서 만든 코드를 가지고 아래의 함수를 생성하시오

    print( cacheProcess( cities, 4 ) )

    결과 : ['jeju', 'pangyo', 'new york', 'new york']

```
cities = ['Jeju', 'Pangyo', 'New York', 'new york']

def cacheProcess( cities, cachesize ):
    city = [ i.lower() for i in cities ]   # cities 값들을 소문자로 변경
    cache = [ None for i in range(1,cachesize+1) ]   # None 4개로 cache 구성
    for i in range(0, len(city)):    # 0, 1, 2, 3
        cache.append( city[i] )
        del cache[0]
    return cache

print( cacheProcess( cities, 4 ) )
```

**■ 파이썬으로 합집합과 교집합 구현하기**

```
a = { 1, 2, 3, 4 }
b = { 2, 4, 5 }
print(type(a))    # <class 'set'> : 집합 자료구조
```

** 파이썬의 자료유형
 1. 리스트
 2. 사전형(딕서너리형)
 3. 튜플형
 4. 집합형

```
a = { 1, 2, 3, 4 }
b = { 2, 4, 5 }

# 1. a 와 b 의 합집합 구하기  ( print(a&b)도 가능 )
result1 = a.union(b)
print(result1)     # {1, 2, 3, 4, 5}

# 2. a 와 b 의 교집합 구하기  ( print(a|b)도 가능 )
result2 = a.intersection(b)
print(result2)     # {2, 4}

# 3. a 와 b 의자 카드 유사도를 구하기
result3 = len(result2)/len(result1)
print(result3)     # 0.4
```

**■ 리스트로 합집합과 교집합 구하기**

```
a = [ 1, 2, 3, 4 ]
b = [ 2, 4, 5 ]

# 1. a 와 b 의 합집합 구하기
result1 = set(a+b)
print(result1)   # {1, 2, 3, 4, 5}

# 2. a 와 b 의 교집합 구하기
a = [ 1, 2, 3, 4 ]
b = [ 2, 4, 5 ]
result2 = []
for i in a:   # 개수가 많은 리스트가 for 문에
    if i in b:
        result2.append(i)
print(result2)   # [2, 4]
```
